```python
#Question 3
class Marsupial:
    def __init__(self):
        self.pouch = []
        
    def put_in_pouch(self, item):
        self.pouch.append(item)
        
    def pouch_content(self):
        return self.pouch
class Kangaroo(Marsupial):
    def __init__(self, x=0, y=0):
        super().__init__()
        self.x = x
        self.y = y
        
    def jump(self, dx, dy):
        self.x += dx
        self.y += dy
        
    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"
k = Kangaroo(0, 0)
print(k)

k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_content())

k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

```python
#Question 6 
from html.parser import HTMLParser


class Headingparser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.indent_level = 0

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h'):
            print(' ' * self.indent_level, end='')
            print(self.get_starttag_text())
            self.indent_level += 1

    def handle_endtag(self, tag):
        if tag.startswith('h'):
            self.indent_level -= 1

    def handle_data(self, data):
        data = data.strip()
        if data and self.indent_level > 0:
            print(' ' * self.indent_level + data)


infile = open('w3c.txt')
content = infile.read()
infile.close()


parser = Headingparser()
parser.feed(content)
```

```python
#Question 7
from urllib.request import urlopen
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup

def webdir(url, depth, indent=0):
    print(" " * indent + url)

    if depth == 0:
        return

    parts = urlparse(url)
    base_url = f"{parts.scheme}://{parts.netloc}"
    try:
        response = urlopen(url)
    except:
        return
    html = response.read().decode("utf-8")

    soup = BeautifulSoup(html, "html.parser")

    for link in soup.find_all("a"):
        href = link.get("href")
        if href is not None and href.startswith("http"):
            new_url = href
        else:
            new_url = urljoin(base_url, href)
        webdir(new_url, depth - 1, indent + 4)

webdir("https://en.wikipedia.org/wiki/Cricket", 2, 0)
```

```python
#Question 8
import sqlite3

conn = sqlite3.connect('Climate_Data.db')
conn.execute('CREATE TABLE Climate_data (City text, Country text, Season text, Temperature real, Rainfall real)')

conn.executemany('INSERT INTO Climate_data VALUES (?, ?, ?, ?, ?)', [
    ('Mumbai', 'India', 'Winter', 24.8, 5.9),
    ('Mumbai', 'India', 'Spring', 28.4, 16.2),
    ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
    ('Mumbai', 'India', 'Fall', 27.6, 346.0),
    ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
    ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
    ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
    ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
    ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
    ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
    ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
    ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)
])
conn.commit()

print(30* '_')
print('Output 1-All the temperature data')
result = conn.execute('SELECT Temperature FROM Climate_data')
print(list(result))

print(30* '_')
print('Output 2- All the cities, but without repetition')
result = conn.execute('SELECT DISTINCT City FROM Climate_data')
print(list(result))

print(30* '_')
print('Output 3- All the records for India.')
result = conn.execute('SELECT * FROM Climate_data WHERE Country = "India"')
print(list(result))

print(30* '_')
print('Output 4-  All the Fall records')
result = conn.execute('SELECT * FROM Climate_data WHERE Season = "Fall"')
print(list(result))

print(30* '_')
print('Output 5-The city, country, and season for which the average rainfall is between 200 and 400 millimeters')
result = conn.execute('SELECT City, Country, Season FROM Climate_data WHERE Rainfall BETWEEN 200 AND 400')
print(list(result))
      
print(30* '_')
print('Output 6- The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.')
result = conn.execute('SELECT City, Country FROM Climate_data WHERE Season = "Fall" GROUP BY City, Country HAVING AVG(Temperature) > 20 ORDER BY AVG(Temperature) ASC')
print(list(result))
      
print(30* '_')
print('Output 7-The total annual rainfall for Cairo.')
result = conn.execute('SELECT SUM(Rainfall) FROM Climate_data WHERE City = "Cairo"')
print(list(result))
      
print(30* '_')
print('Output 8-The total rainfall for each season.')
result = conn.execute('SELECT Season, SUM(Rainfall) FROM Climate_data GROUP BY Season')
print(list(result))
```

```python
#Question 9
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

print(30* '_')
print('Output 1')
Uppercase_words = [word.upper() for word in words]
print(Uppercase_words)

print(30* '_')
print('Output 2')
Lowercase_words = [word.lower() for word in words]
print(Lowercase_words)


print(30* '_')
print('Output 3')
Word_Lengths = [len(word) for word in words]
print(Word_Lengths)


print(30* '_')
print('Output 4')
Word_info = [[word.upper(), word.lower(), len(word)] for word in words]
print(Word_info)


print(30* '_')
print('Output 5')
Long_Words = [word for word in words if len(word) >= 4]
print(Long_Words)
```
